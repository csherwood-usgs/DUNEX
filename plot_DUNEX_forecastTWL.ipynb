```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

# path to met station data
pathn = r'C:\\crs\\proj\\2019_DUNEX\\Pea_Island_Data\\09_TWL_forecasts\\'
# met station
fn = 'peaisland_twl.csv'

df = pd.read_csv(pathn+fn)
df['dt'] = pd.to_datetime(df['forecastTime'], format='%Y-%m-%d %H:%M:%S.%f')
df
```

```python
blues = ['#eff3ff','#bdd7e7','#6baed6','#3182bd','#08519c']
blue3 = ['#deebf7','#9ecae1','#3182bd']
dstart = datetime.date(2021, 9, 1)
dend = datetime.date(2021, 11, 1)

fig,ax = plt.subplots(1,1,figsize=(15,4))
dt = df['dt'].values
twl = df['twl'].values
twl05 = df['twl05'].values
twl95 = df['twl95'].values

plt.plot(dt,twl05,c=blue3[0],zorder=0)
plt.plot(dt,twl95,c=blue3[0],zorder=1)
plt.plot(dt,twl,c=blue3[2],zorder=3)
plt.xlim([dstart, dend])
_ = plt.ylabel('TWL (m)',fontsize=14)
```
